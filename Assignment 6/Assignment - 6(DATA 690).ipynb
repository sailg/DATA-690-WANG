# We need to change the header of the dataframe such that the column names becomes the header

```python
df1.columns = df1.iloc[0]
df1 = df1.drop(df1.index[0])
```

# Creating a function named cleanse_year() that takes a string as input and removes any asterisks in the string and return the 
cleansed string.

```python
def cleanser_year(str1):
    str2 = str1.replace("*","")
    return str2
```

Testing

```python
cleanser_year("1989*")
```

```python
cleanser_year("20*10")
```

# Applying clense_year() to "Census Year" column

```python
df1["Census Year"] = df1["Census Year"].apply(cleanse_year)
```

# Converting "Census Year" column and "Total Population"to int type

```python
df1["Census Year"] = df1["Census Year"].astype(int)
```

```python
df1["Total Population"] = df1["Total Population"].astype(int)
```

# Creating a function named clense_cost to replace Billion with nine 0's and remove "$" sign from any string passed to it.

```python
def cleanse_cost(str1):
    str1 = str1.replace(" Billion","000000000")
    str1 = str1.replace("$","")
    str1 = str1.replace(",","")
    return str1
    
```

# Testing claense_cost function

```python
cleanse_cost("$13 Billion")
```

# Applying cleanse_cost() to "Census Cost" column

```python
df1["Census Cost"] = df1["Census Cost"].apply(cleanse_cost)
```

```python
df1["Census Cost"] = df1["Census Cost"].astype(float)
```

```python
def cleanse_Avgcost(str1):
    if "cents" in str1:
        str1 = str1.replace(" cents","")
        str1 = float(str1)/100
    else:
        str1 = str1.replace("$","")
    return (str1)
    
```

# Testing clense_Avgcost function.

# Applying cleanse_Avgcost function to "Average Cost Per Person" column

```python
df1["Average Cost Per Person"] = df1["Average Cost Per Person"].apply(cleanse_Avgcost)
```

```python
df1["Average Cost Per Person"] = df1["Average Cost Per Person"].astype(float)
```
